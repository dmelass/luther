```python
## Dara Elass
## Luther MVP

%matplotlib inline
import csv
import pandas as pd
import statsmodels.formula.api as sm
from sklearn.linear_model import LinearRegression
import scipy, scipy.stats
import numpy as np
import matplotlib.pyplot as plt
import random
import math
```

```python
movies = pd.read_csv('action_movies.csv')
movies.head()
```

```python
gross_avg = []
grosses = np.asarray(movies['Actor Grosses'])
for gross in grosses:
    if gross == '[]':
        gross_avg.append(0)
    else:
        gross = gross.replace('[','').replace(']','').split(',')
        gross = [int(x) for x in gross]
        gross_avg.append(np.mean(gross))

movies['Average Actor Grosses'] = gross_avg
movies.head()
```

```python
release_years = []
for i in range(len(movies)):
    release_years.append(int(np.asarray(movies['Release Date'])[i].split(',')[1]))

movies['Release Year'] = release_years
movies.head()
```

```python
action_movies = movies[-np.isnan(movies.Budget)]
print len(action_movies)
action_movies = action_movies[-np.isnan(action_movies['Opening Weekend'])]
print len(action_movies)
action_movies = action_movies[-np.isnan(action_movies['Director Gross'])]
print len(action_movies)
action_movies = action_movies[action_movies['Actor Grosses']!=0]
print len(action_movies)
action_movies.head()
```

```python
training_set_end_year = 2010
training_set = action_movies[action_movies['Release Year'] <= training_set_end_year]
test_set = action_movies[action_movies['Release Year'] > training_set_end_year]
```

```python
# scatter plot of budget and opening weekend domestic gross
budget = training_set['Budget']
opening_wknd = training_set['Opening Weekend']
plt.scatter(budget,opening_wknd)
plt.xlabel('Budget')
plt.ylabel('Opening Weekend Domestic Gross')
plt.title('Opening Wknd Domestic Gross vs. Budget')
```

```python
# scatter plot of director gross and opening weekend domestic gross
dir_gross = training_set['Director Gross']
opening_wknd = training_set['Opening Weekend']
plt.scatter(dir_gross,opening_wknd)
plt.xlabel('Director Average Gross')
plt.ylabel('Opening Weekend Domestic Gross')
plt.title('Opening Wknd Domestic Gross vs. Director Gross')
```

```python
# scatter plot of actor gross and opening weekend domestic gross
actor_gross = training_set['Average Actor Grosses']
opening_wknd = training_set['Opening Weekend']
plt.scatter(actor_gross,opening_wknd)
plt.xlabel('Actors Average Grosses')
plt.ylabel('Opening Weekend Domestic Gross')
plt.title('Opening Wknd Domestic Gross vs. Actor Grosses')
```

```python
# create linear model
training_set['Ones']=1
X = training_set[['Ones','Budget','Director Gross','Average Actor Grosses']]
Y = opening_wknd
model = sm.OLS(Y,X).fit()
model.summary()
```
