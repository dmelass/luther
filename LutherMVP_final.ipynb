### Import modules

```python
import csv
import pandas as pd
import statsmodels.formula.api as sm
from sklearn.linear_model import LinearRegression
import scipy, scipy.stats
import numpy as np
import matplotlib.pyplot as plt
import random
import math
from statsmodels.sandbox.regression.predstd import wls_prediction_std
import plotters
import interactivenamepopper as inp
import sklearn.metrics as met
```

### Load data

```python
movies = pd.read_csv('action_movies_2.csv')
movies.head()
```

### Calculate average actor grosses

```python
gross_avg = []
grosses = np.asarray(movies['Actor Grosses'])
for gross in grosses:
    if gross == '[]':
        gross_avg.append(0)
    else:
        gross = gross.replace('[','').replace(']','').split(',')
        gross = [int(x) for x in gross]
        gross_avg.append(np.mean(gross))

movies['Average Actor Grosses'] = gross_avg
movies.head()
```

### Get Release Year

```python
release_years = []
for i in range(len(movies)):
    release_years.append(int(np.asarray(movies['Release Date'])[i].split(',')[1]))

movies['Release Year'] = release_years
movies.head()
```

### Remove NAN and 0 values

```python
action_movies = movies[-np.isnan(movies.Budget)]
print len(action_movies)
action_movies = action_movies[-np.isnan(action_movies['Opening Weekend'])]
print len(action_movies)
action_movies = action_movies[-np.isnan(action_movies['Director Gross'])]
print len(action_movies)
action_movies = action_movies[action_movies['Average Actor Grosses']!=0]
print len(action_movies)
action_movies[action_movies['Actors']=="[]"].head()
```

### Create training and set set

```python
training_set_end_year = 2008
training_set = action_movies[action_movies['Release Year'] <= training_set_end_year]
test_set = action_movies[action_movies['Release Year'] > training_set_end_year]
print len(training_set)
print len(test_set)
```

### Create linear model

```python
# create linear model
budget = training_set['Budget']
opening_wknd = training_set['Opening Weekend']
training_set['Ones'] = 1
X = training_set[['Ones','Budget','Director Gross','Average Actor Grosses']]
Y = opening_wknd
model = sm.OLS(Y,X).fit()
model.summary()
```

### Create plots

```python
#variance and MSE
var_opening_wknd = training_set['Opening Weekend'].var()
MSE_X_budget = met.mean_squared_error(training_set['Opening Weekend'],model_budget.predict(X_budget))
MSE_X = met.mean_squared_error(training_set['Opening Weekend'],model.predict(X))

print var_opening_wknd
print MSE_X_budget
print MSE_X
```

```python
#standard deviation and sqrt of MSE
std_opening_wknd = training_set['Opening Weekend'].std()
MSE_X_budget_sqrt = math.sqrt(met.mean_squared_error(training_set['Opening Weekend'],model_budget.predict(X_budget)))
MSE_X_sqrt = math.sqrt(met.mean_squared_error(training_set['Opening Weekend'],model.predict(X)))

print std_opening_wknd
print MSE_X_budget_sqrt
print MSE_X_sqrt
```
