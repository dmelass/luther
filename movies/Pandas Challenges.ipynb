```python
# Dara Elass

%matplotlib inline
import csv
import pickle
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates
from datetime import datetime
import operator

movie_data = csv.DictReader(open('movies.csv','r'))
movies = []
for row in movie_data:
    movies.append(row)

df = pd.DataFrame(movies)
df
```

```python
movie_df = pd.read_csv('movies.csv',parse_dates = {'datetime':['release date']})
time = movie_df['datetime'].astype(datetime)
dtg = movie_df['domestic_total_gross']

plt.plot_date(time,dtg,'gx')
plt.xticks(rotation=45)
plt.xlabel('Date')
plt.ylabel('Domestic Total Gross ($, x 100m)')
```

```python
#Challenge 2: runtime vs domestic total gross
runtime = movie_df['runtime']
plt.plot(runtime,dtg,'bx')
plt.xlabel('Runtime (mins)')
plt.ylabel('Domestic Total Gross ($, x 100m)')
```

```python
# Challenge 3: group your data by Rating and find the average runtime and domestic total gross at each level of Rating
grouped_ratings = movie_df.groupby('rating')

dtg_groups = []
release_date_groups = []
ratings_groups = []
for rating, values in grouped_ratings:
    ratings_groups.append(rating)
    dtg_groups.append(values['domestic_total_gross'])
    release_date_groups.append(values['datetime'])
    print 'The average runtime for rating '+ rating +' is ' +str(sum(values['runtime'])/len(values['runtime']))+' minutes'
    print 'The average domestic total gross for rating '+rating+' is $'+str(sum(values['domestic_total_gross'])/len(values['domestic_total_gross']))
    
```

```python
# Challenge 4: make one figure with (N=the number of MPAA ratings there are) subplots, 
# and in each plot the release date vs the domestic total gross

N=len(grouped_ratings)

# set up subplots
fig = plt.figure(figsize =(15,10))
ax1 = fig.add_subplot(221)
plt.xticks(rotation=45)
ax2 = fig.add_subplot(222, sharex=ax1, sharey=ax1)
plt.xticks(rotation=45)
ax3 = fig.add_subplot(223, sharex=ax1, sharey=ax1)
plt.xticks(rotation=45)
ax4 = fig.add_subplot(224, sharex=ax1, sharey=ax1)
plt.xticks(rotation=45)
fig.subplots_adjust(hspace=.5)

# release date vs domestic total gross
the_subplots = [ax1,ax2,ax3,ax4]
for i in range(N):
    the_subplots[i].plot(release_date_groups[i].astype(datetime), dtg_groups[i],'bo')
    the_subplots[i].set_title('Rating: '+ratings_groups[i])
```

```python
# Challenge 5: what director in your dataset has the highest gross per movie?
grouped_dirs = movie_df.groupby('director')

max_dtg = 0
dir_and_grosses = {}
for director, values in grouped_dirs:
    num_movies = len(values['domestic_total_gross'])
    current_dtg = values['domestic_total_gross']
    gross_per_movie = float(sum(current_dtg))/num_movies
    dir_and_grosses[director]=gross_per_movie
   
max_director = max(dir_and_grosses.iteritems(), key=operator.itemgetter(1))[0]   
print 'The director with the highest gross per movie is',max_director
```

```python
# Challenge 6: bin your dataset into years (if applicable) and make a bar graph with error bars of gross each year
months = []
for date in movie_df['datetime']:
    months.append(date.month)

movie_df['months'] = months
grouped_months = movie_df.groupby('months')
means = grouped_months['domestic_total_gross'].mean()
errors = grouped_months['domestic_total_gross'].std()
fix, ax = plt.subplots()
means.plot(yerr=errors, ax=ax, kind='bar')
```
